# Introduction to Data Visualization with Matplotlib

## Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate random number
number = 12

x = np.arange(1, number + 1)

np.random.seed(0)


y1, y2 = np.random.normal(10, 10, size=([2,number]))
```


`fig`
This variable represents the entire figure or the top-level container that holds all the elements of the plot. 
It can be thought of as the canvas on which you will create your visualizations. 
The fig variable is an instance of the Figure class in Matplotlib.

`ax`
This variable represents the axes of a subplot. 
In Matplotlib, an "axes" is an individual plot within the figure. 
Even if you have a single plot, it is still referred to as an "axes." 
The ax variable is an instance of the Axes class in Matplotlib.

```python
fig, ax = plt.subplots()
ax.plot(x, y1)
```

## Customizing your plots

Doc

- [matplotlib.axes.Axes.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html)
- [List of named colors](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
fig, ax = plt.subplots()
# add markers, linstyle, color(or c)
ax.plot(x, y1, 
        marker = 'o', 
        linestyle = '--',
        c = '#0984e3'
       )

# set label
ax.set_xlabel('Time (months)')
ax.set_ylabel('Profit ($ thousand)')
ax.set_title('Profit of Candy Store (2023)')
```

## Small multiples

```python
# create subpolts in 3 rows 2 columns
fig, ax = plt.subplots(3,2)
```

```python
# crete a 2 rows x 1 column subplots where all the sub plots share the same y axis
fig, ax = plt.subplots(2,1, sharey= True)
ax[0].plot(x, y1, c = '#EAB543', marker = 'o', linestyle = '--')
ax[1].plot(x, y2, c = '#1B9CFC', marker = 'o', linestyle = '--')

ax[0].set_title('Profit of Candy Store (2023)')
ax[0].set_ylabel('Profit of store A')
ax[1].set_ylabel('Profit of store B')
ax[1].set_xlabel('Time (months)')

ax[0].plot()
```

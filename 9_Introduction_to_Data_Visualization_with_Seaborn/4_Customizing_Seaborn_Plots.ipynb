# 4 Customizing Seaborn Plots

## Changing plot style and color

- Reasons to change style:

    - Personal preference
    - Improve readability
    - Guide interpretation

Figure "style" includes background and axes

Preset options: "white" , "dark" , "whitegrid" , "darkgrid" , "ticks"

`sns.setstyle()`

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex'
)
```

```python
sns.set_style("whitegrid")

sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex'
)
```

```python
sns.set_style("darkgrid")

sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex'
)
```

## Changing the palette

Figure "palee" changes the color of the main elements of the plot 

`sns.set_palette()` # default Diverging palettes

Use preset palees or create a custom palee

Offcial Doc: [Choosing color palettes](https://seaborn.pydata.org/tutorial/color_palettes.html)

```python
sns.catplot(
    x = 'day',
    data = tips,
    kind = 'count',
)
```

```python
sns.set_style("darkgrid")

sns.set_palette("RdBu")

sns.catplot(
    x = 'day',
    data = tips,
    kind = 'count',
)
```

### Custom Palette

```python
custom_palette = ["#9c88ff", "green" , "orange" , "#00a8ff" , "yellow" , "purple"]
sns.set_palette(custom_palette)
```

```python
sns.catplot(
    x = 'day',
    data = tips,
    kind = 'count',
)
```

## Change the Scale

Figure "context" changes the scale of the plot elements and labels

`sns.set_context()`

Smallest to largest: "paper", "notebook", "talk", "poster"

```python
sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex'
)
```

```python
sns.set_context('talk')

sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex'
)
```

## Adding titles and labels

FacetGrid vs. AxesSubplot objects

Seaborn plots create two dierent types of objects: `FacetGrid` and `AxesSubplot`



### Adding a title to FacetGrid

```python
g = sns.catplot(
    x = 'day',
    data = tips,
    kind = 'count',
)

g.fig.suptitle(
    'Title',
    y = 1.03
)
g.tick_params(labelrotation=90)
```

### Adding a title to AxesSubPlot

```python
g = sns.countplot(
    x = 'day',
    data = tips,
)

g.set_title(
    'Title',
    y = 1.03
)
```

```python
g = sns.catplot(
    x = 'day',
    data = tips,
    kind = 'count',
    col = 'time'
)

g.fig.suptitle(
    'FacetGrid Title',
    y = 1.03
)
g.set_titles('AxesSubPlot for {col_name}')

# x-axis
g.set(
    xlabel = 'X label',
    ylabel = 'Y label'
)

# Rotating x-axis tick
g.tick_params(labelrotation=90)
```

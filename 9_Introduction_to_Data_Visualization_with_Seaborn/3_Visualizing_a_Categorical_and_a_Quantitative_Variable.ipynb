# Visualizing a Categorical and a Quantitative Variable

## Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')
```

```python
sns.set_theme()
```

```python
tips.info()
```

```python
order = [
    'Thur',
    'Fri',
    'Sat',
    'Sun'
]

sns.catplot(
    x = 'day',
    data = tips,
    kind = 'count',
    hue = 'time',
    col = 'smoker',
    order = order
)
```

## Bar Plot

Displays *mean of quantitative variable* per category

```python
order = [
    'Thur',
    'Fri',
    'Sat',
    'Sun'
]

sns.catplot(
    x = 'day',
    y = 'tip',
    data = tips,
    kind = 'bar',
    hue = 'time',
    order = order
)
```

## Box plot

```python
order = [
    'Dinner',
    'Lunch',
]

sns.catplot(
    x = 'time',
    y = 'total_bill',
    data = tips,
    kind = 'box',
    # remove outlier
    # sym = ''
    # By default, the whisker is 1.5x IQR
    # we can custom it by set it to 2.0, [5, 95], [0,100](min&max -> no outlier)
    whis = [5, 95]
)
```

## Pint Plot

- Points show mean of quantitative variable- Vertical lines show 95% con,dence intervals

Point plots vs. line plots

Both show:

Mean of quantitative variable

95% con,dence intervals for the mean

Differences:
Line plot has *quantitative* variable (usually time) on x-axis
Point plot has *categorical* variable on x-axis

```python
sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'bar',
    hue = 'sex'
)
```

```python
sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex'
)
```

Advantages for point plot.

It's easier to compare the heights of the subgroup points when they're stacked above each other. 

In the point plot, it's also easier to look at the differences in slope between the categories than it is to compare the heights of the bars between them

```python
from numpy import median

# median is more robust to outliers

sns.catplot(
    x = 'time',
    y = 'tip',
    data = tips,
    kind = 'point',
    hue = 'sex',
    estimator = median,
    # add caps to the end of the confidence intervals
    capsize = 0.1,
    # remove the lines joining the points
    join = False
)
```

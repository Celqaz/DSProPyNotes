# 2. Visualizing Two Quantitative Variables

## Introduction to relational plots and subplots

Why use `relplot()` instead of `scatterplot()`?

`relplot()` lets you create subplots in a single figure

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = sns.load_dataset('tips')
```

```python
tips.info()
```

### Subplots in columns

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'smoker'
)
```

### Subplots in rows

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    row = 'smoker'
)
```

### Subplots in rows and columns

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    row = 'smoker'
)
```

### Wrapping columns

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    col_wrap = 2
)
```

```python
# Ordering columns
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    col = 'day',
    col_wrap = 2,
    col_order=["Thur","Fri","Sat","Sun"]
)
```

## Customizing scatter plots

### size, hue, alpham

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    size = 'size',
    hue = 'size',
    alpha = 0.6
)
```

### style

```python
sns.relplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    kind = 'scatter',
    hue = 'smoker',
    style = 'smoker',
)
```

## Introduction to line plots

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

mpg = sns.load_dataset('mpg')
```

```python
mpg.info()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            errorbar=None, style="origin", 
            hue="origin", markers = True, dashes = False)

# Show plot
plt.show()
```

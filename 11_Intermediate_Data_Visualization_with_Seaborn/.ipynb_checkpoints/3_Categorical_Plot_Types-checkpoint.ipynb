# 3 Categorical Plot Types

- stripplot
- swarmplot

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
data = sns.load_dataset('mpg')
```

```python
data.info()
```

```python
data['cylinders'].value_counts()
```

## Stripplot

```python
sns.stripplot(
    data = data,
    y = 'origin',
    x = 'mpg',
    jitter = True
)
```

## Swarmplot

```python
sns.swarmplot(
    data = data,
    y = 'origin',
    x = 'mpg',
    #  change the size in seaborn scatterplot
    s = 3
)
```

## Boxplot

```python
sns.boxplot(
    data = data,
    y = 'origin',
    x = 'mpg',
)
```

## Violinplot

 - suitable for large datasers
 - computionally intensive 

```python
sns.violinplot(
    data = data,
    y = 'origin',
    x = 'mpg',
)
```

## Boxenplot

 - hybird for boxplot and violinplot
 - esay to generate

```python
sns.boxenplot(
    data = data,
    y = 'origin',
    x = 'mpg',
)
```

## Barplot

```python
sns.barplot(
    data = data,
    y = 'origin',
    x = 'mpg',
    hue = 'cylinders'
)
```

## Pointplot

```python
sns.pointplot(
    data = data,
    x = 'mpg',
    y = 'origin',
    hue = 'cylinders'
)
```

## Countplot

```python
sns.countplot(
    data = data,
    x = 'origin',
    hue = 'cylinders'
)
```

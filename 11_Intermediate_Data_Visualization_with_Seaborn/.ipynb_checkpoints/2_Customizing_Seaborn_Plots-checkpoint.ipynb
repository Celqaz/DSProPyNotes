# Customizing Seaborn Plots

```python
import seaborn as sns
import matplotlib
```

```python
tips = sns.load_dataset('tips')
```

## Using Seaborn Styles

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# apply default seaborn style
sns.set()
```

```python
# in pandas
tips['total_bill'].plot.hist()
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex')
```

```python
sns.histplot(tips, x="tip", hue="time", element="step", kde = True)
```

```python
sns.set_style('white')
sns.histplot(tips, x="tip", hue="time", element="step", kde = True)
```

```python
matplotlib.rc_file_defaults()
```

```python
sns.displot(x = 'total_bill', data = tips)
# remove left axes
sns.despine(left = True)
```

## Colors in Seaborn

```python
# activate color codes
sns.set(color_codes = True)
sns.histplot(tips, x="tip", element="step", color = 'r')
```

```python
palettes = ['deep', 'muted', 'pastel', 'bright','dark','colorblind']
for p in palettes:
    sns.set_palette(p)
    # sns.color_palette() : get current palette
    # sns.palplot(): display pallette
    sns.palplot(sns.color_palette())
sns.displot(tips, x="tip", hue = 'time', element= 'step')
```

## Customizing Data

```python
# Circular colors = when the data is notordered
sns.palplot(sns.color_palette("Paired", 12))
```

```python
# Sequential colors = when the data has aconsistent range from high to low
sns.palplot(sns.color_palette("Blues", 12))
```

```python
# Diverging colors = when both the low andhigh values are interesting
sns.palplot(sns.color_palette("BrBG", 12))
```

## Customizing with matplotlib

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
sns.histplot(tips['total_bill'], ax = ax)
ax.set(
    xlabel = 'Total Bill($)',
    ylabel = 'Distribution',
    title = 'Total Bill Distribution',
    xlim = (0,50)
)
```

multiple plots

```python
fig, (ax0, ax1) = plt.subplots(
    nrows = 1,
    ncols = 2,
    sharey = True,
    figsize = (7,4)
)
sns.histplot(tips['total_bill'], stat = 'density', ax = ax0)
sns.histplot(tips['tip'], stat = 'density', ax = ax1)

ax1.set(xlabel='Tips', xlim=(0, 10))
ax1.axvline(x=4, label='Ecpected Tip', linestyle='--')
ax1.legend()
```
